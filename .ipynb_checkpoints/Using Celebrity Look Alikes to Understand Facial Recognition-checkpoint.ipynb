## What is enrolling images to create a Facial Recognition Database?

It may not be immediately obvious, but in order to find a celebrity a given person looks alike, one needs to create a facial recognition database containing the celebrities one cares to match against non-celebrities. Once the database is created, one can present it with any face of anybody and with a bit of code one can tell which celebrity in the database that anybody-face looks the most alike. 

To create a facial recognition database a process of *enrolling* each face image is performed. *Enrolling* entails collecting some information describing a face image, and then passing both to a pre-trained facial recognition algorithm, in this case DLib's RESNET v1. 

The facial recognition algorithm analyzes the provided data and returns a 128 dimensional vector representing a "point" where that face lands in an imaginary 128 dimensional space. (More about that imaginary 128 dimensional space in a moment.) That 128 dimensional vector is, data-wise, just an array of numbers. For simplicity that array will be called a *faceDescriptor* in this notebook. The *facial recognition industry* calls them *templates*. 

The fact that a *faceDescriptor* is a *point in an imaginary 128 dimensional space* is key to how facial recognition operates. Given two points, their relationship can be described with a single number: the distance between them. When a *facial recognition algorithm* is *trained* that is a process of analyzing millions of face images. However, that analysis includes the additional key information of *which images from the collection are of the same person*. In the collection of images, each person has multiple images, and in many cases dozens of images. By knowing which images are of the same face, the facial algorithm training can find facial characteristics that persist across variations of view. 

When training is completed, the trained algorithm can convert a face image into a 128 dimensional point. Face images from the same person will land near one another, while face images from other people land far away in this imaginary 128 dimensional space. 

In order to tell if a photo is of a given person in a facial recognition database, one takes the unknown person's photo, converts it to a RESNET 128 dimensional point, and then locates the closest 128 dimensional point in the facial recognition database. If that distance is beneath a threshold, then the unknown image is considered to be identified as that of the person from the database with the least distance. 

Exactly the threshold distance one uses as *close* varies with *facial recognition algorithm*. But the basic concept remains the same: some abstract higher dimensional space is used to locate a face as a point in that abstract space, and the distance between two "face points" can be used to verify if the images are of the same person.

Getting back to our **Celebrity Look Alike Finder**, what is needed is to enroll all the celebrity photos, acquiring their 128 dimensional points, and save them. The collection of 128 dimensional points, and who each point belongs is the *facial recognition database*. When looking for a celebrity look alike, get the non-celebrity's 128 dimensional point and simply find which celebrity has the closest 128 dimensional point. 

The cell beneath enrolls our celebrity photos, creating the *facial recognition database*. The cell below may require a half hour or more to run, depending upon your system. On a Ryzen 7 3700X 8-Core Processor this takes 30 minutes. A min info report prints when completed. The code is commented for your understanding: 

```python
# variables for enrollment:
faceDescriptors = None  # will be the enrolled faces descriptors when finished
name_index = {}         # for each face descriptor, this is the celeb name
path_index = {}         # for each face descriptor, path to the enrollment image of face descriptor
#
success_count = 0       
error_count = 0
image_count = 0
celeb_count = 0

enroll_start = time.time()  # remember this moment in time

# looping over each subfolder:
for folder in subfolders:
    celeb_count += 1
    
    # dereference to get our celeb name:
    celeb_name = labelMap[folder]
    
    # get list of images in this folder:
    image_files = os.listdir(os.path.join(faceDatasetFolder, folder))
    
    # looping over all files in this folder:
    for file in image_files:
        image_count += 1
        
        # get the full path to this specific file:
        full_path = os.path.join(faceDatasetFolder,folder,file)
        
        # things happen and non-image files sometimes end up in image directories:
        if not full_path.lower().endswith(('.bmp', '.jpg', '.jpeg', '.jpe', '.png')):
            continue
            
        print( "{} via file {} being processed...".format( celeb_name, full_path ))
        
        # read image, arrives in BGR_8-bits-per-pixel format
        im_bgr8 = cv2.imread(full_path)
        
        # need image in RGB format:
        im_rgb8 = cv2.cvtColor(im_bgr8, cv2.COLOR_BGR2RGB)
        
        # detect faces in image
        face_rects = faceDetector(im_rgb8, 0)
        
        face_count = len(face_rects)
        if face_count<1:
            print("failed to find face in {}!".format(full_path))
            error_count += 1
        elif face_count>1:
            print("found multiple faces in {}".format(full_path))
            # choose the face with the largest area:
            maxArea = 0
            maxRect = None
            for face in face_rects:
                if face.area() > maxArea:
                    maxArea = face.area()
                    maxRect = [face.left(), face.top(), face.right(), face.bottom()]
                    
            # put data in same format as if only one face were in that image:
            rect = dlib.rectangle(*maxRect)
            face_rects = [rect]
            
        # back from above if/else, now we only care about face_rects of length 1:
        face_count = len(face_rects)
        if face_count==1:
            
            # have DLib find/predict the facial landmarks:
            dlibFaceShape = shapePredictor(im_rgb8, face_rects[0])
            
            # convert landmarks from Dlib's format to list of (x, y) points
            landmarks = [(p.x, p.y) for p in dlibFaceShape.parts()]
            
            lmark_count = len(landmarks)
            if lmark_count!=68:
                print("only found {} landmarks in {}".format(lmark_count, full_path))
                error_count += 1
            else:
                # Compute face descriptor using DLib's resnet model v1 neural network.
                # A Face Desciptor is a 128D vector that describes a face in an image 
                # identified by a DLib FaceShape:
                faceDescriptor = faceRecognizer.compute_face_descriptor(im_rgb8, dlibFaceShape)
                
                # Convert face descriptor from Dlib's format to list, then a NumPy array
                faceDescriptorList = [x for x in faceDescriptor]
                faceDescriptorNdarray = np.asarray(faceDescriptorList, dtype=np.float64)
                faceDescriptorNdarray = faceDescriptorNdarray[np.newaxis, :]
                
                # Stack face descriptors (1x128) for each face in images, as rows
                if faceDescriptors is None:
                    faceDescriptors = faceDescriptorNdarray
                else:
                    faceDescriptors = np.concatenate((faceDescriptors, faceDescriptorNdarray), axis=0)
                    
                # remember this success-number-as-index to the celeb name, parallel with faceDescriptors
                name_index[success_count] = celeb_name
                path_index[success_count] = full_path
                success_count += 1

# finished looping over the enrollments:
enroll_end = time.time()
enroll_time = enroll_end - enroll_start

print( "We've processed {} images for {} celebrities in {} seconds.".format(image_count, celeb_count, enroll_time))

print( "errors {}".format(error_count))
print( "successes {}".format(success_count))

# Write face descriptors and index to disk
np.save('descriptors.npy', faceDescriptors)
# name_index has celeb names in same order as descriptors in faceDescriptors
with open('name_index.pkl', 'wb') as f:
    cPickle.dump(name_index, f)
# path_index has image paths in same order as descriptors in faceDescriptors
with open('path_index.pkl', 'wb') as f:
    cPickle.dump(path_index, f)
    
```

## Briefly revisiting the enrollment process to understand what's next
The Python code that enrolls all the celebrity photos is two nested loops. <br>
- The outside loop goes through each celebrity subfolder one at a time. 
  - The inside loop starts with one a folder and loops through each photo in the folder. <br>
    The work within the inside loop is the enrollment process for each image:

1. Load the image, it arrives in BGR pixel format
1. Convert image to RGB pixel format, DLib expects RGB
1. Detect faces in image, receive array of face rects
   1. If no faces, skip image
   1. If more than one face, choose the largest face
1. Use face rect with image to get image's face landmarks
   1. if face landmark count is not 68, skip image
1. Use face landmarks with image to get 128 dimensional vector *faceDescriptor*
1. Once a face image is verified to be usable and the *faceDescriptor* is acquired, each *faceDescriptor* (just an array of 128 numbers), the celebrity name corresponding and the path to the enrollment photo are stacked into their appropriate parallel array. 

### How finding a Celebrity Look Alike is *Almost* Facial Recognition

In order to find a celebrity look alike, steps 1 through 5 are repeated with the new non-celebrity photo. Step 5 produces the *faceDescriptor* of the non-celebrity photo. That *faceDescriptor* is a 128 dimensional point in some imaginary mathmatical space. The *reason* it is such an odd thing was engineered on purpose to make the next step very easy: just calculate the distance between the new *faceDescriptor* and the array of *faceDescriptors* composing the celebrity face database. The closest celebrity faceDescriptor with the non-celebrity faceDescriptor are look alikes - at least as close as is contained in the face database. 

The difference between this and *real facial recognition* is the calculated distance between the non-celebrity photo and the celebrity's matching photo. In *real facial recognition* the distance is very close, while a look alike has no such requirement. A look alike is just the celebrity with the least distance to a non-celebrity.

Below is the final Python that repeats steps 1 through 5, then performs the distance calculation with the *facial recognition database*, locates the closest one and shows the results. This will read all the jpeg images from inside the *test-images* directory and display their look alikes. Take notice of the distances generated between any non-celebrity photos and their counterparts, a final discussion follows. 

```python
matplotlib.rcParams['figure.figsize'] = (12.0,12.0) # set up the plotting for output display
```

```python
# get image filenames:
testImages = glob.glob('test-images/*.jpg')

# loop over each:
for test in testImages:
    
    # read from disk:
    im = cv2.imread(test)
    
    # convert to RGB pixel format:
    imDlib = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    
    # detect faces in image
    face_rects = faceDetector(imDlib, 0)
    
    face_count = len(face_rects)
    if face_count!=1:
        print("Problem with test image {}!".format(test))
    else:
        # Find facial landmarks
        dlibFaceShape = shapePredictor(imDlib, face_rects[0])
        
        # convert landmarks from Dlib's format to list of (x, y) points
        landmarks = [(p.x, p.y) for p in dlibFaceShape.parts()]
        
        lmark_count = len(landmarks)
        if lmark_count!=68:
            print("Problem with test image {}!".format(test))
        else:
            # Compute face descriptor using neural network defined in Dlib.
            # It is a 128D vector that describes the face in imDlib identified by dlibFaceShape.
            faceDescriptor = faceRecognizer.compute_face_descriptor(imDlib, dlibFaceShape)
            
            # Convert face descriptor from Dlib's format to list, then a NumPy array
            faceDescriptorList = [x for x in faceDescriptor]
            faceDescriptorNdarray = np.asarray(faceDescriptorList, dtype=np.float64)
            faceDescriptorNdarray = faceDescriptorNdarray[np.newaxis, :]
            
            # Calculate Euclidean distances between face descriptor just calculated
            # against all the face descriptors calculated while enrolling faces:
            distances = np.linalg.norm(faceDescriptors - faceDescriptorNdarray, axis=1)
            
            # Calculate minimum distance and index of this face:
            argmin = np.argmin(distances)    # index
            minDistance = distances[argmin]  # minimum distance
            #
            # this is our celeb doppleganger's name:
            doppleganger_name = name_index[argmin]
            #
            # this is the path to our doppleganger image:
            doppleganger_path = path_index[argmin]
    
    celeb_name = doppleganger_name
    ####################
    
    plt.subplot(121)
    plt.imshow(imDlib)
    plt.title("test img")
    
    # display celeb image which looks closest to the test image: 
    plt.subplot(122)
    doppleganger_im = cv2.imread(doppleganger_path)
    doppleganger_imDlib = cv2.cvtColor(doppleganger_im, cv2.COLOR_BGR2RGB)
    plt.imshow(doppleganger_imDlib)
    plt.title("Celeb Look-Alike={}, Distance={:1.2}".format(celeb_name,minDistance))
    plt.show()
    
```

Did you take note of the distance between each test image and the celebrity photo? <br>

Now try adding photos to the *test-images* directory, trying to see if you can find a photo producing a distance greater than 0.8. It is hard to find a photo above 0.8 that is not a actually a photo of a celebrity in the database. It is quite difficult. That is the **artificial intelligence**. <br>

The *deep learning training* produces an algorithm, in this case *DLib's RESNET v1*, that found the unique characteristics of a human face which persist regardless of the quality of the face image. <br>

In the specific case of *DLib's RESNET v1* it produces a 128 number array, representing a 128 dimensional point in an imaginary space. <br>

To compare two face photos, a distance between each 128 dimensional point produced by each photo is compared. If that distance is *very close* the photos are very likely of the same person. Pretty nifty. 

## Summary

The above discussion and Python code should give one a deeper understanding of *facial recognition* and by extension both *deep learning* and *artificial intelligence*.  <br>

In the case of *facial recognition* it is a process of using *deep learning* to find the human facial features that persist for an individual despite common variations of appearance.  <br>

*Deep Learning* produces a *trained algorithm* that, in this instance, given face images produces points in an imaginary high dimensional space, which will only be *high dimensionally close* if the face images are of the same person. <br>

*Face recognition* is the process of collecting and enrolling faces of interest, and then comparing the high dimensional points their images' produce. If the points are close and the *trained algorithm* is accurate, the photos will be of the same person.  <br>

The *threshold minimum distance* used to verify two photos are of the same person varies with each *trained algorithm*.

Now, from what has been discussed, the following additional information can be derived:

- The *faceDescriptors* from different *trained algorithms* are incompatible. This is due to the plain fact that *different training image sets* are used to create *different trained algorithms*. <br>


- The contents of the multi-million image set used in *deep learning* to train a *facial recognition algorithm* is vitally important to contain as many variations of face images as possible. <br>


- The *trained algorithm* produced needs to be trained with wide variations and combinations of angle, expression, illumination, zoom, ethnicity, age, gender, partial occlusions as well as qualities of image encoding in order to find the persistent facial features that last *despite all these variations*. <br>


- Ideally, an image training set contains dozens if not hundreds of images of each unique individual, across environments, night and day, variations of weather, emotion, hair cuts, makeups and/or beards and anything else one might vary in the ordinary course of how humans vary their appearance. Such an image data set should produce is high quality *face recognition system*. 

A final point of interest: DLib's RESNET v1 was trained with a data set of 3 million faces. This notebook's  __[author](https://github.com/bsenftner)__ spent seven years working on three generations of a __[facial recognition system](https://cyberextruder.com/)__ trained with over 700 million faces.
